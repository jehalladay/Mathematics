```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def first_order_ode(f, y0, t):
    y = np.zeros(len(t))
    y[0] = y0
    for n in range(0,len(t)-1):
        y[n+1] = y[n] + f(y[n],t[n])*(t[n+1] - t[n])
    return y
```

```python
def second_order_ode(f, y0, dy0, t):
    y = np.zeros(len(t))
    dt = (t[1] - t[0])
    y[0] = y0
    y[1] = dy0*dt+y0
    for n in range(0,len(t)-2):
        y[n+2] = (2 * y[n+1]) - y[n] + f(y[n], (y[n+1]-y[n])/dt, t[n])*(dt**2)
        'y[n+2] = y[n+1] + .5*H(3f(y[n+1], (y[n-1]-y[n])/dt, t[n]) - f(y[n], (y[n-1]-y[n])/dt, t[n]))'
    return y
```

```python
g = lambda t: t**2/2+t+3/2
f = lambda y, t: t+1
```

```python
t=np.linspace(1,11,60)
y0 = 3
y = first_order_ode(f, y0, t)
y_true = g(t)
plt.plot(t,y,'b.-',t,y_true,'r-')
plt.legend(['Euler','True'])
plt.axis([0,12,0,80])
plt.grid(True)
plt.title("Solution of $y'=x+1 , y(1)=3$")
plt.show()
```
